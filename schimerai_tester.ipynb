```python
from river import datasets
import schimerai
import pandas as pd
import numpy as np
```

```python
bananas = datasets.Bananas() # y is what we predict and we append x to it for our input
```

```python
prev_input = 0
```

```python
agent_count = int(100) #arbitary
extra_data_agent_count = int(9) #arbitary
prev_pred_diff = 0
W = pd.DataFrame(np.ones((agent_count, agent_count)))
L = pd.DataFrame(np.ones((agent_count, agent_count)))
P = pd.DataFrame(np.zeros((agent_count, agent_count)))
AV = pd.DataFrame(np.ones((agent_count, 1)))

w = 1.0

curr_state = {
    'agent_count': agent_count,
    'extra_data_agent_count': extra_data_agent_count,
    'prev_pred_diff': prev_pred_diff,
    'W': pd.DataFrame.to_dict(W),
    'L': pd.DataFrame.to_dict(L),
    'P': pd.DataFrame.to_dict(P),
    'AV': pd.DataFrame.to_dict(AV)
}
```

```python
for x,y in bananas:
    input = [int(i) for i in x.values()]
    input.insert(0,y)
    print(curr_state)
    output, curr_state = schimerai.schimerai(list(input), curr_state)
    prev_input = output
```
